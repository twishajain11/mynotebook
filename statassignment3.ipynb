Question 1: What is hypothesis testing in statistics?
Answer:
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It starts with formulating two competing statements: the null hypothesis (H₀), which represents a default or no-effect assumption, and the alternative hypothesis (H₁ or Ha), which represents the effect or relationship we are testing for. Using sample data, we calculate a test statistic and compare it against a critical value or p-value to decide whether to reject or fail to reject the null hypothesis. Hypothesis testing helps determine if observed patterns or differences in data are statistically significant or likely due to random chance, making it essential in research, business decisions, and scientific studies.

Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?
Answer:
In statistics, the null hypothesis (H₀) is a statement that assumes no effect, no difference, or no relationship exists in the population. It represents the default position that any observed differences in sample data are due to random chance. The alternative hypothesis (H₁ or Ha), on the other hand, is the statement that assumes there is an effect, difference, or relationship. Essentially, while the null hypothesis assumes nothing interesting is happening, the alternative hypothesis represents the outcome the researcher expects or wants to test. During hypothesis testing, the goal is to use sample data to decide whether there is enough evidence to reject H₀ in favor of H₁. For example, if a company wants to test whether a new advertising strategy increases sales, H₀ would state that the strategy has no impact, whereas H₁ would state that it does increase sales.

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
Answer:
In hypothesis testing, the significance level (denoted by α) is the threshold used to determine whether the observed data is unlikely under the null hypothesis (H₀). It represents the probability of making a Type I error, which occurs when we incorrectly reject a true null hypothesis. An extremely significant level refers to a very small α value (e.g., 0.01 or 0.001), meaning we require very strong evidence from the data before rejecting H₀. The smaller the significance level, the more stringent the test, and the less likely we are to falsely claim that an effect exists. In practice, if the p-value calculated from the data is less than α, we reject H₀ and conclude that the results are statistically significant; otherwise, we fail to reject H₀. Using an extremely low significance level ensures that only very strong evidence leads to rejecting the null, reducing the risk of false positives.

Question 4: What are Type I and Type II errors? Give examples of each.
Answer:
In hypothesis testing, errors can occur because decisions are based on sample data rather than the entire population. A Type I error occurs when we reject the null hypothesis (H₀) even though it is true, essentially a false positive. For example, if a company tests a new drug and concludes that it works when, in reality, it has no effect, that is a Type I error. A Type II error occurs when we fail to reject the null hypothesis even though the alternative hypothesis (H₁) is true, essentially a false negative. For instance, if the same drug actually works but the test concludes that it does not, that is a Type II error. Understanding these errors is important because they help researchers balance the risk of false positives and false negatives when designing experiments and choosing significance levels.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.
Answer:
A Z-test and a T-test are both statistical methods used to determine whether there is a significant difference between sample data and a population or between two samples. The Z-test is used when the population standard deviation is known and the sample size is large (usually n > 30), relying on the normal distribution. In contrast, the T-test is used when the population standard deviation is unknown and/or the sample size is small (usually n ≤ 30), relying on the t-distribution, which accounts for extra variability in smaller samples. For example, if a company wants to test whether the average weight of items produced matches the known population standard, a Z-test is appropriate. However, if the standard deviation is unknown and only a small sample of items is measured, a T-test should be used. In short, the choice depends on knowledge of the population standard deviation and the size of the sample.

```python
"""Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]"""
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


population_mean = 50
population_std = 0.8  

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
"""Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib"""
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)  # for reproducibility
mu = 50             # population mean
sigma = 5           # population standard deviation
sample_size = 100

data = np.random.normal(mu, sigma, sample_size)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  
standard_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
z_score = stats.norm.ppf(0.975)  
ci_lower = sample_mean - z_score * standard_error
ci_upper = sample_mean + z_score * standard_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval: [{:.2f}, {:.2f}]".format(ci_lower, ci_upper))

plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dotted', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dotted', linewidth=2, label='95% CI Upper')
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
"""Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean."""
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1) 
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6]

z_scores = calculate_z_scores(data)
print("Z-scores:", z_scores)

plt.hist(z_scores, bins=8, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
plt.legend()
plt.show()
```

A Z-score measures how many standard deviations a data point is from the mean. It is calculated as:
Z = \frac{X - \mu}{\sigma}
X = individual data point
μ = mean of the dataset
σ = standard deviation of the dataset

Z-scores standardize the data, making it easier to compare values from different datasets or detect outliers. A Z-score of 0 means the value equals the mean, positive Z means above the mean, and negative Z means below the mean.
