Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).
Answer:
1. Artificial Intelligence (AI):
AI is the broadest concept. It refers to machines or systems that can perform tasks that normally require human intelligence, like understanding language, recognizing images, or making decisions.
Example: Chatbots, self-driving cars, and smart assistants.

2. Machine Learning (ML):
ML is a subset of AI. It focuses on training machines to learn patterns from data and improve automatically without being explicitly programmed.
Example: Predicting house prices based on past sales data.
3. Deep Learning (DL):
DL is a subset of ML. It uses artificial neural networks with many layers (hence “deep”) to learn from large amounts of complex data.
Example: Image recognition, voice assistants, and translation apps.
Key point: Deep learning needs a lot of data and computational power.

4. Data Science:
Data Science is the field of extracting knowledge and insights from data. It uses statistics, ML, and sometimes AI/deep learning to analyze data and make decisions.
Example: Analyzing customer data to improve sales or detect fraud.

Question 2: What are the types of machine learning? Describe each with one
real-world example.
Answer:
Types of Machine Learning and Examples:
1. Supervised Learning:
The model is trained on labeled data (input-output pairs). It learns to predict the output for new inputs.
Example: Predicting house prices based on features like size and location. The model learns from past houses with known prices.

2. Unsupervised Learning:
The model is trained on unlabeled data and tries to find patterns or structure in the data.
Example: Customer segmentation in marketing — grouping customers with similar behavior without knowing any labels in advance.

3. Semi-Supervised Learning:
Uses a small amount of labeled data and a large amount of unlabeled data to improve learning accuracy.
Example: Medical image analysis — only a few images are labeled by experts, but the model learns from thousands of unlabeled images too.

4. Reinforcement Learning (RL):
The model learns by interacting with the environment and getting rewards or penalties based on its actions.
Example: Training a self-driving car — it learns to drive safely by receiving feedback for good or bad actions.

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.
Answer:
Overfitting
Occurs when a machine learning model learns the training data too well, including noise and random fluctuations.
The model performs very well on training data but poorly on new, unseen data.
Example: A model predicting house prices memorizes every tiny detail of the training data instead of learning general patterns.

Underfitting
Happens when a model is too simple to capture the underlying pattern in the data.
It performs poorly on both training and test data.
Example: Trying to fit a straight line to data that has a curved trend.

Bias
Bias is the error due to overly simplistic assumptions in the model.
High bias → model is too simple → underfitting.
Example: Assuming all houses have the same price regardless of size.

Variance
Variance is the error due to the model being too sensitive to training data.
High variance → model is too complex → overfitting.
Example: Model changes drastically if you slightly change the training dataset.

Bias-Variance Tradeoff
There’s a balance between bias and variance.
Goal: Choose a model that captures the patterns (low bias) but does not memorize noise (low variance).
Too simple → high bias, low variance → underfitting
Too complex → low bias, high variance → overfitting

Visual: Think of it like a target board:
High bias → arrows are all far from the center (systematically wrong)
High variance → arrows scatter wildly (inconsistent)
Low bias & low variance → arrows cluster near the center (best model)

Question 4: What are outliers in a dataset, and list three common techniques for
handling them.

Outliers in a Dataset
Outliers are data points that are significantly different from the rest of the data.
They can be unusually high or low values that don’t follow the general pattern.
Outliers can occur due to measurement errors, data entry mistakes, or natural variation.
Example: In a dataset of students’ ages, if most are between 15–18 but one entry is 50, that’s an outlier.

Three Common Techniques to Handle Outliers
1. Removing Outliers:
Simply delete the extreme values from the dataset.
Use this when the outlier is clearly an error and doesn’t represent reality.

2. Transforming Data:
Apply transformations like log, square root, or Box-Cox to reduce the impact of extreme values.

3. Replacing or Capping (Winsorization):
Replace outliers with a percentile value (like 1st or 99th percentile) to reduce their effect without losing data.

Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.
Answer:
Handling Missing Values
Missing values occur when some data points are not recorded or are empty in a dataset. Handling them is important because they can lead to inaccurate analysis or errors in machine learning models.
Steps to handle missing values:

1. Identify missing values – check which columns or rows have missing data.
2. Decide on a strategy – depending on the type of data and the amount of missingness.
3. Handle missing values – using techniques like deletion or imputation.

Common Imputation Techniques:
Numerical Data:
Mean/Median Imputation – replace missing values with the mean or median of the column.
Example: If some students’ exam scores are missing, replace them with the average score.
Categorical Data:
Mode Imputation – replace missing values with the most frequent category.
Example: If gender is missing in some rows, replace it with the most common gender in the dataset.

```python
#Question 8: Write a Python script to:
#● Generate 1000 samples from a normal distribution.
#● Introduce 50 random missing values.
#● Fill missing values with the column mean.
#● Plot a histogram before and after imputation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame(data, columns=['Value'])

# 2. Introduce 50 random missing values
missing_indices = np.random.choice(df.index, 50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# 3. Plot histogram before imputation
plt.figure(figsize=(10, 4))
plt.hist(df['Value'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 4. Fill missing values with the column mean
mean_value = df['Value'].mean()
df['Value'].fillna(mean_value, inplace=True)

# 5. Plot histogram after imputation
plt.figure(figsize=(10, 4))
plt.hist(df['Value'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Histogram After Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

print(df.head())
```

```python
# Question 10: You are working as a data scientist for a retail company. You receive a customertransaction dataset that contains:● Missing ages,● Outliers in transaction amount,● A highly imbalanced target (fraud vs. non-fraud),● Categorical variables like payment method.Explain the step-by-step data preparation plan you’d follow before training a machine learningmodel. Include how you’d address missing data, outliers, imbalance, and encoding.

"""
Step-by-Step Data Preparation Plan
-----------------------------------

1️⃣ Handle Missing Data (e.g., missing ages)
   - Use mean/median imputation for numerical columns.
   - For categorical columns, use the most frequent value.

2️⃣ Handle Outliers (e.g., unusual transaction amounts)
   - Use the Interquartile Range (IQR) method to detect and cap outliers.

3️⃣ Handle Imbalanced Target (fraud vs. non-fraud)
   - Use SMOTE (Synthetic Minority Oversampling Technique) to balance the dataset.

4️⃣ Handle Categorical Variables (e.g., payment method)
   - Use One-Hot Encoding for nominal categorical variables.
   - Use Label Encoding for ordinal variables if any.

5️⃣ Final Step
   - Split the data into training and testing sets.
   - Scale the features if required (e.g., using StandardScaler).
"""

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Example dataset
data = {
    'age': [25, np.nan, 40, 22, np.nan, 35, 60],
    'transaction_amount': [100, 2000, 150, 120, 5000, 90, 130],
    'payment_method': ['Credit Card', 'Cash', 'UPI', 'Credit Card', 'Cash', 'UPI', 'Credit Card'],
    'is_fraud': [0, 0, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# 1️⃣ Handle Missing Values
imputer = SimpleImputer(strategy='mean')
df['age'] = imputer.fit_transform(df[['age']])

# 2️⃣ Handle Outliers (IQR method)
Q1 = df['transaction_amount'].quantile(0.25)
Q3 = df['transaction_amount'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df['transaction_amount'] = np.where(df['transaction_amount'] > upper, upper,
                                    np.where(df['transaction_amount'] < lower, lower, df['transaction_amount']))

# 3️⃣ Encode Categorical Variables
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded = encoder.fit_transform(df[['payment_method']])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['payment_method']))
df = pd.concat([df.drop('payment_method', axis=1), encoded_df], axis=1)

# 4️⃣ Handle Imbalance (Manual Oversampling)
fraud = df[df['is_fraud'] == 1]
non_fraud = df[df['is_fraud'] == 0]

fraud_oversampled = fraud.sample(len(non_fraud), replace=True, random_state=42)
df_balanced = pd.concat([non_fraud, fraud_oversampled]).sample(frac=1, random_state=42)

# 5️⃣ Scale the Features
scaler = StandardScaler()
X = df_balanced.drop('is_fraud', axis=1)
y = df_balanced['is_fraud']
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
print("\nAfter Data Preparation:\n", X_scaled_df)
print("\nBalanced Target Distribution:\n", y.value_counts())
```
