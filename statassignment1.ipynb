Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
Answer:
Difference Between Descriptive and Inferential Statistics

Statistics is the branch of mathematics that deals with collecting, organizing, analyzing, and interpreting data.
It is mainly divided into two branches — Descriptive Statistics and Inferential Statistics.
Both are equally important, but they serve different purposes in data analysis.

1. Descriptive Statistics
Definition:
Descriptive statistics refers to methods used for summarizing, organizing, and describing the main features of a dataset.
It helps us to present data in a meaningful and understandable form.
However, it does not allow us to make conclusions or predictions beyond the data that has been collected.
Purpose:
To describe what the data shows. It gives a snapshot of the data’s central tendency, dispersion, and overall pattern.

Common Tools and Measures:
Measures of Central Tendency: Mean, Median, Mode
Measures of Dispersion: Range, Variance, Standard Deviation
Graphs and Charts: Bar graphs, Pie charts, Histograms, Frequency tables

Example 1 :
Suppose a teacher records the marks of 50 students in a math exam.
She calculates:
Mean (average) score = 75
Highest score = 98
Lowest score = 40
Standard deviation = 10
These statistics help her describe the performance of her class — how students scored overall, and how much variation there was in their marks.
But she cannot say how all students in the school performed, because this data is only about one class.


2. Inferential Statistics
Definition:
Inferential statistics refers to methods used to make predictions, estimates, or draw conclusions about a population based on a sample of data.
It uses probability theory to help determine how reliable those conclusions are.
Purpose:
To make generalizations beyond the available data — in other words, to infer information about a large group from a small sample.
Common Tools and Techniques:
Hypothesis Testing (e.g., t-test, z-test)
Confidence Intervals
Correlation and Regression Analysis
Analysis of Variance (ANOVA)
Chi-square Tests

Example 1 :
A researcher wants to know the average height of 10th-grade students in India.
It’s impossible to measure every student, so she takes a sample of 200 students from different schools.
She finds their average height is 160 cm.
Using inferential statistics, she estimates that the average height of all 10th graders in India is around 160 cm, within a certain margin of error.
Thus, she has made an inference about a population using a sample.


Descriptive statistics deals with what has happened — it describes the data that is already available.
Inferential statistics deals with what might happen — it helps us make decisions or predictions about the larger population using sample data.

Both are essential:
Descriptive statistics gives the foundation and overview.
Inferential statistics helps in decision-making, forecasting, and hypothesis testing.


Question 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.
Answer:

Sampling is the process of selecting a small group (sample) from a larger population to collect data and draw conclusions about the whole population.
Since studying every member of a population is often impossible, expensive, or time-consuming, we use samples that represent the entire group.
Example:
If a school has 2,000 students and you want to find the average height, it’s not practical to measure everyone.
Instead, you measure the heights of 200 students chosen carefully — that group of 200 is your sample, and all 2,000 students make up the population.

Random sampling and stratified sampling are two common methods of selecting a sample from a population. In random sampling, every individual in the population has an equal chance of being selected, and there is no prior division into subgroups. This method is simple and easy to apply and works well when the population is fairly homogeneous, but it may not represent all subgroups proportionally. For example, if a school has 1,000 students and 100 are picked entirely at random, some groups like boys or girls might be over- or under-represented.
                     On the other hand, stratified sampling involves dividing the population into subgroups, called strata, based on a characteristic such as gender, age, or class, and then taking a random sample from each subgroup in proportion to its size. This ensures that all important subgroups are represented and is especially useful for heterogeneous populations. For instance, if the same school has 600 boys and 400 girls and a total of 100 students are to be surveyed, stratified sampling would select 60 boys and 40 girls randomly. While stratified sampling is slightly more complex than random sampling, it provides a more accurate representation of the population. In short, random sampling gives everyone an equal chance but may miss subgroups, whereas stratified sampling ensures proportional representation of all subgroups.

Question 3: Define mean, median, and mode. Explain why these measures of central
tendency are important.
Answer:
Measures of Central Tendency: Mean, Median, and Mode
Definition:
Measures of central tendency are statistical tools used to identify the center or average value around which the entire set of data revolves.
The three main measures are Mean, Median, and Mode.
They help us understand what is “typical” or “normal” in a dataset.

1. Mean (Arithmetic Average)
Definition:
The mean is the sum of all observations divided by the total number of observations.
Formula:
\text{Mean} = \frac{\text{Sum of all observations}}{\text{Total number of observations}}
Example:
If the marks of five students are 60, 70, 80, 90, and 100,
then
\text{Mean} = \frac{60 + 70 + 80 + 90 + 100}{5} = \frac{400}{5} = 80

On average, the students scored 80 marks.

2. Median (Middle Value)

Definition:
The median is the middle value of a dataset when all the observations are arranged in ascending or descending order.
If the number of observations is even, the median is the average of the two middle values.
Example:
Marks: 50, 60, 70, 80, 90
Here, the middle value is 70, so the median is 70.
if the marks were 50, 60, 70, 80, 90, 100,
then the median = (70 + 80) / 2 = 75.
Half of the students scored below 70 and half scored above 70.

3. Mode (Most Frequent Value)
Definition:
The mode is the value that occurs most frequently in a dataset.
A dataset can have:
One mode (unimodal),
Two modes (bimodal), or
More than two modes (multimodal).
Example:
If the test scores are 60, 65, 70, 70, 75, 80,
then the mode = 70, because it appears most often.
The most common score among students is 70.
. Importance of Mean, Median, and Mode

These measures are called measures of central tendency because they help us understand the center or typical value of a dataset.
They are important for several reasons:
1. Simplifies Data:
They summarize a large amount of data with a single representative value.
2. Comparison:
They allow easy comparison between different groups or datasets.
Example: Comparing the average marks of two classes.
3. Decision-Making:
Businesses, teachers, and researchers use these values to make informed decisions.
Example: A company may use the mean income of employees to decide salary ranges.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?
Answer:
1. Skewness
Definition:
Skewness is a statistical measure that describes the degree of asymmetry or departure from symmetry in the distribution of data.
In a perfectly symmetrical distribution, the left and right sides of the graph (such as a histogram or bell curve) are mirror images of each other.
But most real-world data are not perfectly symmetrical, which is where skewness comes in.

Types of Skewness:

A. Positive Skew (Right-Skewed Distribution)
B. Negative Skew (Left-Skewed Distribution)
C. Zero Skew (Symmetrical Distribution)

2. Kurtosis
Definition:
Kurtosis measures the “peakedness” or flatness of a data distribution compared to a normal distribution.
It tells us how heavy or light the tails of the distribution are — in other words, how extreme the outliers are.

Types of Kurtosis:
A. Mesokurtic (Normal Distribution)
B. Leptokurtic (High Kurtosis)

. Positive Skew Implication
A positive skew means:
The data is concentrated toward lower values.
A few unusually high values stretch the tail to the right.
The mean is pulled higher than the median and mode.
Real-world examples include income, house prices, or extreme test scores where most observations are low/moderate but a few are very high.

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Pythonlist_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]
import numpy as np

x = [10, 20, 30, 40, 60]
y = [15, 25, 35, 45, 60]

x_array = np.array(x)
y_array = np.array(y)

cov_matrix = np.cov(x_array, y_array, ddof=1)  
cov_xy = cov_matrix[0, 1]  
corr_matrix = np.corrcoef(x_array, y_array)
corr_xy = corr_matrix[0, 1]  

print("Covariance between x and y:", cov_xy)
print("Correlation coefficient between x and y:", corr_xy)
```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data, vert=True, patch_artist=True, showmeans=True)
plt.title("Box Plot of Given Data")
plt.ylabel("Values")
plt.show()

Q1 = np.percentile(data, 25)  
Q3 = np.percentile(data, 75)  
IQR = Q3 - Q1  


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("First Quartile (Q1):", Q1)
print("Third Quartile (Q3):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers in the data:", outliers)
```

Explanation
1. Box Plot Components:
Bottom line of the box: Q1 (25th percentile)
Top line of the box: Q3 (75th percentile)

Middle line: Median (50th percentile)
Whiskers: Extend to the smallest and largest values within 1.5 × IQR from Q1 and Q3
Points outside whiskers: Outliers

2. IQR Method for Outliers:
Calculate IQR = Q3 - Q1
Lower bound = Q1 - 1.5 × IQR
Upper bound = Q3 + 1.5 × IQR
Any data point outside these bounds is considered an outlier.


```python
"""Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]"""

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

x = np.array(advertising_spend)
y = np.array(daily_sales)

cov_matrix = np.cov(x, y, ddof=1)  # sample covariance
cov_xy = cov_matrix[0, 1]

corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]

print("Covariance between advertising spend and daily sales:", cov_xy)
print("Correlation coefficient between advertising spend and daily sales:", corr_xy)
```

●
To explore the relationship between advertising spend and daily sales, we can use covariance and correlation. Covariance measures how two variables move together; a positive covariance indicates that as advertising spend increases, daily sales tend to increase, while a negative covariance would indicate the opposite. However, covariance alone is difficult to interpret because it is not standardized. The correlation coefficient, on the other hand, quantifies both the strength and direction of the linear relationship between two variables, ranging from -1 (perfect negative correlation) to 1 (perfect positive correlation). A high positive correlation would suggest that increasing advertising spend is strongly associated with higher daily sales, which is useful for the marketing team in planning their budgets.
