```python
import pandas as pd

url = "https://drive.google.com/uc?id=1iKy23bMtEQShF_weneRNnYrFmzvpPOI3"

# If it's a CSV file
df = pd.read_csv(url)

print("First 10 rows:")
print(df.head(10))

print("\nShape of dataset:", df.shape)

print("\nColumn names:")
print(df.columns.tolist())
```

```python
#2 Question 2: Check for missing values in all columns and describe your approach for handling them.(Include your Python code and output in the code box below.)

print("Missing values per column:")
print(df.isnull().sum())

print("\nPercentage of missing values per column:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['selling_price'], bins=50, edgecolor='black')
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.title("Distribution of Bike Selling Prices")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by seller_type and owner (or score type) and calculate mean price
avg_prices = df.groupby(['seller_type', 'owner'])['selling_price'].mean().unstack()

# Plot
avg_prices.plot(kind='bar', figsize=(10,6), edgecolor='black')

plt.title("Average Selling Price by Seller Type and Owner")
plt.xlabel("Seller Type")
plt.ylabel("Average Selling Price")
plt.legend(title="Owner / Score Type")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Compute average km_driven for each ownership type
avg_km = df.groupby('owner')['km_driven'].mean().sort_values()

# Plot
plt.figure(figsize=(10,6))
avg_km.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Average Kilometers Driven by Ownership Type")
plt.xlabel("Ownership Type")
plt.ylabel("Average Kilometers Driven")
plt.xticks(rotation=45)
plt.show()

print("Average km driven for each ownership type:\n")
print(avg_km)
```

```python
import pandas as pd

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# ---- Before removing outliers ----
print("Summary statistics BEFORE outlier removal:\n")
print(df[['km_driven', 'selling_price']].describe())

# ---- Remove outliers from both columns ----
df_no_outliers = remove_outliers_iqr(df, 'km_driven')
df_no_outliers = remove_outliers_iqr(df_no_outliers, 'selling_price')

# ---- After removing outliers ----
print("\nSummary statistics AFTER outlier removal:\n")
print(df_no_outliers[['km_driven', 'selling_price']].describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['year'], df['selling_price'], alpha=0.5, color='teal', edgecolor='k')

plt.title("Year vs Selling Price of Bikes")
plt.xlabel("Year of Manufacture")
plt.ylabel("Selling Price")
plt.show()
```
